```python
#  pip install requests
# pip install BeautifulSoup4
# pip install html5lib

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# create requets to the site you want to scrap

data_source = requests.get("https://www.pararius.com/apartments/amsterdam/page-1").text # getting htm page from the server
data_source2 = requests.get('https://www.pararius.com/apartments/amsterdam/page-2').text
```

```python
# convert the response object to beautiful soup object, we use "html5lib" to parse

soup_data = BeautifulSoup(data_source,'html5lib')
soup_data2 = BeautifulSoup(data_source2,'html5lib')
contents = soup_data.find_all("section",class_="listing-search-item")
contents2= soup_data2.find_all("section",class_="listing-search-item")
```

```python
contents
```

```python
result = [] # create a list that will hold the dictionaries(List of dictionary)

for content in contents:
    data ={} # creating a dictionary
    apart_name = content.find("h2",class_="listing-search-item__title").text.replace('\n','')
    location  = content.find("div", class_="listing-search-item__location").text.replace('\n','')
    price = content.find("div", class_="listing-search-item__price").text.replace('\n','')
    area = content.find("li", class_="illustrated-features__item illustrated-features__item--surface-area").text.replace('\n','')
    data['Apart_name'] = apart_name
    data['Location'] = location
    data['Price'] = price
    data['Area'] = area
    result.append(data)
```

```python
amaster = pd.DataFrame(result)
amaster
```
